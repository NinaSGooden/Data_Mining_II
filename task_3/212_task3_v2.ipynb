```python
#Create lists to prepare datasets to be encoded
lists = []
for i in range (0,7501):
    lists.append([str(df.values[i,j]) for j in range(0,20)])
lists
```

```python
#Transactionalize before using Apriori()
TE = TransactionEncoder()
array = TE.fit(lists).transform(lists)
```

```python
#Create DataFrame of data
cleaned_df = pd.DataFrame(array, columns = TE.columns_)
cleaned_df
```

```python
#Check to see if there are empty columns by searhing for "NaN" values
for col in cleaned_df.columns:
    print(col)
```

```python
#Drop empty columns
df_cleaned = cleaned_df.drop(['nan'], axis = 1)
```

```python
#Check to see if columns were successfully dropped
for col in df_cleaned.columns:
    print(col)
```

```python
df_cleaned.shape
```

```python
df.to_csv('E:/HDD Documents/HDD Documents/School/WGU/d212_data_mining_ii/task_3_df_cleaned.csv')
```

```python
#Run .apriori() to mine rules
a_rules = apriori(df_cleaned, min_support = .05, use_colnames = True)
a_rules.head()
```

```python
#Set metrics and mining rules
ass_r = association_rules(a_rules, metric = 'lift', min_threshold = 1)
ass_r
```

```python
ass_values = ass_r[['antecedents', 'consequents', 'support','confidence','lift']]
ass_values
```

```python
ass_sort = ass_r.sort_values(['lift', 'confidence'], ascending = True)
ass_sort
```

```python
#Deteremind rules based on set conditions
ass_sort[ (ass_sort['lift'] >= 1.3) & (ass_sort['confidence'] >= .25) ]
```
